# Image Seam Speed Contest! (Due october 13th)

A pretty fun demo is the <a href="https://github.com/alanedelman/18.337_2015/blob/master/Lecture04_0921/ImageSeam.ipynb">image seam demo  </a> shown in class in lecture 4.

<br>

1) Warmup: Replace stata with any image of your own choosing, and get the demo to work.  There may be issues
with Images, Color/Colors, and Interact, but there are unlikely to be any issues with the algorithm itself.
If you face issues, please post to Julia google groups or the github issues pages.  Mine work with
Julia 0.3.11 and failed to work with Julia 0.4. <br>

You may need to do Pkg.add("Images") , Pkg.add("Interact"), etc., once.  (The sequence is
Pkg.add once, which downloads the package, and "using" once per session, which locally loads
the package for the session.)
I expect there to be issues here.

<br>

2) Speed contest:  For a Tosci's ice cream certificate, we will have two contests.  The goal is to time
the compression part (the part that says "Stata takes a while to compute,") and get it to go as fast
as possible with Julia on a machine of your choice, and on my mac.  

<br>
Submit your time and your code, and we might do a bake off in class.
<br>

You can use parallelism, a way better algorithm (probably most profitable!), or anything you like.
If you think a GPU can help you should use a Julia GPU package.
